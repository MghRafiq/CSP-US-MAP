## 1. Introduction détaillée

Un **Problème de Satisfaction de Contraintes (CSP)** est une tâche d'optimisation où l'on cherche à attribuer des valeurs à un ensemble de variables tout en respectant un ensemble de contraintes définies. Ce cadre général est utilisé pour modéliser et résoudre une variété de problèmes complexes, allant des puzzles aux applications industrielles.

### Définition formelle

Un CSP est défini par un triplet **(X, D, C)**, où :
- **Variables (X)** : Ensemble des entités à affecter (X = {X₁, X₂, ..., Xₙ}).
- **Domaines (D)** : Ensemble des valeurs possibles pour chaque variable (D = {D₁, D₂, ..., Dₙ}).
- **Contraintes (C)** : Règles qui restreignent les combinaisons autorisées des valeurs (C = {C₁, C₂, ..., Cₘ}).

L'objectif est de trouver une affectation **A : X → D** telle que toutes les contraintes **C** soient satisfaites.

### Exemples d'applications des CSP

1. **Planification et gestion** :
   - Allocation des ressources.
   - Création d'emplois du temps (exemple : planification scolaire).

2. **Théorie des graphes** :
   - Résolution de problèmes de coloriage de graphes.
   - Recherche de chemins dans des réseaux.

3. **Intelligence artificielle** :
   - Résolution de puzzles (exemple : Sudoku).

4. **Recherche Opérationnelle (RO)** :
   - Optimisation combinatoire (problème du sac à dos, voyageur de commerce).


## 2. Représentation mathématique et modélisation

Les CSP peuvent être représentés mathématiquement sous forme de graphes pour mieux structurer le problème.

### Exemple : Colorier une carte

Le problème consiste à attribuer une couleur à chaque région d'une carte de manière à ce que deux régions adjacentes n'aient pas la même couleur. Dans cet exemple :
- **Variables (X)** : {A, B, C, D, E} (chaque région correspond à une variable).
- **Domaines (D)** : {Rouge, Vert, Bleu, Jaune} (les couleurs possibles pour chaque région).
- **Contraintes (C)** : Deux régions adjacentes ne peuvent pas partager la même couleur.

### Visualisation graphique

Le problème peut être représenté comme un graphe G = (V, E), où :
- Les sommets **V** représentent les régions.
- Les arêtes **E** représentent les adjacences entre régions.

### Objectif

Attribuer des couleurs à chaque sommet du graphe tout en respectant les contraintes des arêtes.

## 3. Algorithmes de résolution
Plusieurs approches sont utilisées pour résoudre les CSP, selon leur complexité :

### **Backtracking**

L'algorithme de **backtracking** explore toutes les affectations possibles en vérifiant les contraintes à chaque étape.

#### Étapes de l'algorithme
1. Choisir une variable non assignée.
2. Tester toutes les valeurs possibles dans le domaine de cette variable.
3. Appliquer les contraintes pour réduire les choix des variables restantes.
4. Si une contrainte est violée, revenir en arrière (backtracking).

### **Optimisation avec heuristiques**
  * Minimum Remaining Values (MRV) : Choisir la variable ayant le moins d'options possibles dans son domaine.
  * Forward Checking : Réduire les domaines des variables non affectées après chaque choix.

### **Recherche locale**
  * Hill Climbing : Partir d'une solution partielle et améliorer progressivement.
  * Simulated Annealing : Approche probabiliste pour éviter les minima locaux.

## 4. Étude de cas : Coloration avec 4 couleurs

Le problème des 4 couleurs stipule que toute carte plane peut être colorée avec un maximum de 4 couleurs sans que deux régions adjacentes aient la même couleur.

### **Problème**
Colorier une carte simplifiée avec les régions suivantes et leurs adjacences :
  * Régions : A, B, C, D, E.
  * Adjacences : (A, B),(A, C)(B,C),(B, D),(C, D),(D, E).
  
### **Résolution**
  * Variables : {A, B, C, D, E}.
  * Domaines : {Rouge, Vert, Bleu, Jaune}.
  * Contraintes : Les couleurs des deux régions adjacentes doivent être différentes.
**Solution possible**
A = Rouge, B = Vert, C = Bleu, D = Rouge, E = Vert.
### **Étapes de la solution :**
  1. Définir les variables et les domaines :
    * **Variables (X) :** Chaque région de la carte est une variable, donc X={A, B, C, D, E}.
    * **Domaines (D) :** Les couleurs disponibles pour chaque région sont D={Rouge, Vert, Bleu, Jaune}.

  2. Identifier les contraintes :
    * Les contraintes sont que deux régions adjacentes ne peuvent pas avoir la même couleur. Par exemple, les régions A et B ne doivent pas partager la même couleur, A et C ne doivent pas partager la même couleur, etc.
    * Cela se traduit par les contraintes suivantes :
      * A ≠ B
      * A ≠ C
      * B ≠ C
      * B ≠ D
      * C ≠ D
      * D ≠ E

  3. Choisir une variable à affecter (choix de la première région à colorier) :
    * Par convention, on commence par colorier une région quelconque. Ici, nous commencerons avec la région A.
    * Affectation : On affecte la couleur "Rouge" à A, c'est-à-dire A = Rouge.

  4. Appliquer les contraintes pour réduire le domaine des autres variables :
    * Une fois que la région A est colorée en "Rouge", les régions adjacentes à A (B et C) ne peuvent pas être "Rouge".
    * Cela réduit les domaines de B et C à {Vert, Bleu, Jaune}.

  5. Choisir la prochaine variable à affecter (région B) et une couleur possible :
    * Choisissons la région B et affectons-lui la couleur "Vert" (car "Rouge" est déjà pris par A).
    * Affectation : B = Vert.

  6. Appliquer les contraintes et réduire les domaines :
    * Maintenant, B est colorée en "Vert", donc les régions adjacentes à B (A et D et C) ne peuvent pas être "Vert".
    * Cela réduit le domaine de D à {Rouge, Bleu, Jaune} (car A = Rouge ) et le domaine de C à {Bleu, Jaune} (car A = Rouge et B = Vert).

  7. Colorier la région C :
    * Choisissons la région C, qui est adjacente à A et B , et affectons-lui la couleur "Bleu" (car "Rouge" est déjà pris par A et  "Vert" est pris par B ).
    * Affectation : C = Bleu.

  8. Réduire le domaine de la région D :
    * Maintenant que C = Bleu et B = Vert, la région D, qui est adjacente à B et C, ne peut pas être "Vert" ni "Bleu".
    * Cela réduit le domaine de D à {Rouge, Jaune}.

  9. Colorier la région D :
    * Choisissons la région D, et affectons-lui la couleur "Rouge" (car "Vert" et "Bleu" sont déjà utilisés pour les autres régions adjacentes).
    * Affectation : D = Rouge.

  10. Colorier la région E :
    * Enfin, colorions la région E, qui est adjacente à D mais E n'est pas adjacente à C, B, ou A . Étant donné que D = Rouge, E peut être colorée avec n'importe quelle couleur qui n'a pas encore été utilisée par D, car la contrainte ne concerne que les régions adjacentes, E peut être colorée en Vert, Bleu, ou Jaune.
    * Affectation : E = Vert.

````python

# Import des bibliothèques nécessaires
import networkx as nx
import matplotlib.pyplot as plt
from typing import List, Dict

# Classe pour résoudre le problème de satisfaction de contraintes (PSC)
class CSP:
    def __init__(self, variables: List[str], domains: Dict[str, List[str]], constraints: Dict[str, List[str]]):
        """
        Initialiser un problème de satisfaction de contraintes (PSC).
        :param variables: Liste des variables (régions à colorier).
        :param domains: Dictionnaire associant chaque variable à son domaine de valeurs possibles.
        :param constraints: Dictionnaire des contraintes (adjacences des régions).
        """
        self.variables = variables
        self.domains = domains
        self.constraints = constraints

    def is_valid_assignment(self, variable: str, value: str, assignment: Dict[str, str]) -> bool:
        """
        Vérifie si une affectation est valide selon les contraintes.
        """
        for neighbor in self.constraints[variable]:
            if neighbor in assignment and assignment[neighbor] == value:
                return False
        return True

    def backtrack(self, assignment: Dict[str, str] = {}) -> Dict[str, str]:
        """
        Résolution du PSC par backtracking.
        """
        if len(assignment) == len(self.variables):  # Si toutes les variables sont affectées
            return assignment

        unassigned = [var for var in self.variables if var not in assignment]
        current_var = unassigned[0]  # Choix de la première variable libre

        for value in self.domains[current_var]:
            if self.is_valid_assignment(current_var, value, assignment):
                assignment[current_var] = value
                result = self.backtrack(assignment)
                if result:  # Si une solution est trouvée
                    return result
                del assignment[current_var]  # Backtracking
        return None  # Aucune solution trouvée


# Fonction pour afficher le graphe et ses couleurs à chaque étape
def plot_graph(graph, coloring=None, title=""):
    pos = nx.spring_layout(graph)  # Positionner les nœuds
    node_colors = [coloring[node] if coloring and node in coloring else "white" for node in graph.nodes]
    nx.draw(graph, pos, with_labels=True, node_color=node_colors, node_size=800, font_size=10)
    plt.title(title)
    plt.show()


# Définir les régions, couleurs et contraintes
variables = ['A', 'B', 'C', 'D', 'E']
domains = {var: ['red', 'green', 'blue', 'yellow'] for var in variables}
constraints = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D'],
    'D': ['B', 'C', 'E'],
    'E': ['D']
}

# Création d'une instance CSP
csp = CSP(variables, domains, constraints)

# Résolution du problème
solution = csp.backtrack()

# Création du graphe des régions
graph = nx.Graph()
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E')]
graph.add_nodes_from(variables)
graph.add_edges_from(edges)

# Visualiser les étapes
plot_graph(graph, title="Graphe des régions et contraintes (Initial)")

if solution:
    # Affichage progressif des colorations
    plot_graph(graph, title="Étape 1 : Aucune région colorée")
    partial_coloring = {}

    # Ajouter et afficher chaque coloration étape par étape
    for region, color in solution.items():
        partial_coloring[region] = color
        plot_graph(graph, coloring=partial_coloring, title=f"Coloration progressive : Région {region} colorée en {color}")

    # Affichage final
    plot_graph(graph, coloring=solution, title="Résultat final : Graphe coloré")
else:
    print("Aucune solution trouvée.")
````

## 6. Complexité des Problèmes de Satisfaction de Contraintes (CSP)

Un Problème de Satisfaction de Contraintes (CSP) consiste à assigner des valeurs à un ensemble de variables tout en respectant des contraintes spécifiques. La complexité de ces problèmes réside dans le nombre potentiellement immense de combinaisons à explorer pour trouver une solution valide.

### Complexité brute :

- **Variables (n)** : Nombre total de variables impliquées dans le CSP.
- **Domaines (d)** : Taille moyenne du domaine de chaque variable, c'est-à-dire le nombre de valeurs possibles pour chaque variable.

Dans le pire des cas, l'espace de recherche peut contenir jusqu'à \( d^n \) combinaisons possibles d'assignations. Cette croissance exponentielle rend la résolution directe des CSP particulièrement coûteuse en termes de temps de calcul.

### Exemples de complexité :

- **Coloriage de graphe** : Pour un problème de coloriage avec 5 régions et 4 couleurs, il existe \( 4^5 = 1\,024 \) combinaisons possibles à examiner sans tenir compte des contraintes d'adjacence.
- **Sudoku classique** : Avec 81 cases et 9 valeurs possibles par case, l'espace total des combinaisons est de \( 9^{81} \), un nombre astronomique rendant une approche exhaustive impraticable.

### Approches plus efficaces que les CSP traditionnels :

Bien que les CSP offrent un cadre puissant pour modéliser et résoudre des problèmes combinatoires, certaines méthodes peuvent s'avérer plus efficaces selon le contexte :

- **Programmation Linéaire (PL)** :
  * *Applicabilité* : Lorsque les contraintes et la fonction objectif sont linéaires.
  * *Efficacité* : Des algorithmes tels que le simplexe ou les méthodes de points intérieurs permettent de résoudre ces problèmes en temps polynomial, souvent en \( O(n^3) \) pour les méthodes de points intérieurs.
  * *Exemple* : Optimisation de l'utilisation des ressources dans un planning avec des contraintes linéaires.

- **Recherche locale** :
  * *Applicabilité* : Pour des problèmes où une solution optimale exacte n'est pas impérative.
  * *Efficacité* : Des méthodes comme le Hill Climbing ou le Recuit Simulé explorent localement l'espace des solutions, avec une complexité souvent réduite à \( O(n) \) ou \( O(n^2) \), et peuvent converger rapidement vers une solution satisfaisante.
  * *Exemple* : Planification de tâches dans un environnement dynamique.

- **Méthodes SAT (Satisfaction booléenne)** :
  * *Applicabilité* : Lorsque les contraintes peuvent être exprimées sous forme de formules booléennes.
  * *Efficacité* : Bien que la complexité puisse atteindre \( O(2^n) \) dans le pire des cas, des techniques d'optimisation comme la propagation unitaire et l'élagage rendent les solveurs SAT plus performants que les CSP traditionnels pour certains problèmes spécifiques.

### Pourquoi utiliser les CSP ?

Les CSP sont particulièrement adaptés lorsque :

- Les contraintes sont complexes et potentiellement non linéaires.
- Le problème combine des relations logiques et arithmétiques.
- Une exploration exhaustive est nécessaire pour garantir une solution optimale.

En somme, bien que les CSP offrent un cadre général pour la résolution de problèmes combinatoires complexes, l'utilisation d'approches alternatives peut, dans certains cas, fournir des solutions plus efficaces en termes de temps de calcul et de ressources.

### 8. Choix de l'algorithme pour la coloration de la carte des États-Unis et relation avec les problèmes de satisfaction de contraintes (CSP)

### Complexité du théorème des quatre couleurs :
Bien que le théorème des quatre couleurs affirme qu'il est possible de colorier n'importe quelle carte plane avec seulement quatre couleurs, sans que deux régions adjacentes partagent la même couleur, il ne fournit pas de méthode pratique pour déterminer cette coloration. De plus, certaines configurations géographiques, comme des États partageant de longues frontières ou des points de contact complexes, rendent l'application pratique du théorème ardue sans un algorithme efficace.

### Algorithme DSATUR :
 Pour surmonter ces défis, l'algorithme DSATUR (Degree of Saturation) est particulièrement adapté. Cet algorithme colore les sommets d'un graphe en choisissant à chaque étape le sommet non colorié ayant le degré de saturation le plus élevé, c'est-à-dire celui adjacent au plus grand nombre de couleurs différentes déjà utilisées. En cas d'égalité, il sélectionne le sommet avec le degré le plus élevé parmi les non coloriés. Cette approche permet de minimiser le nombre de couleurs utilisées tout en évitant les conflits entre régions adjacentes.

### Coloration de graphes en tant que CSP :
La problématique de la coloration de graphes peut être formulée comme un problème de satisfaction de contraintes (CSP). Dans ce cadre :

* Variables : Chaque région ou État est représenté par une variable.
* Domaines : Chaque variable peut prendre une valeur parmi un ensemble de couleurs disponibles.
* Contraintes : Deux variables correspondant à des régions adjacentes ne doivent pas avoir la même valeur (couleur).
### Avantages de la modélisation CSP :

* Flexibilité : Permet d'incorporer facilement des contraintes supplémentaires ou spécifiques au problème étudié.
* Utilisation d'algorithmes génériques : Les solveurs de CSP peuvent être appliqués directement pour trouver une solution ou déterminer l'insatisfiabilité du problème.

### Limites :

* Complexité computationnelle : Certains problèmes de coloration de graphes, lorsqu'ils sont modélisés en tant que CSP, peuvent être NP-difficiles, rendant leur résolution computationnellement intensive.
* Scalabilité : Pour les grands graphes, la modélisation en tant que CSP peut entraîner une explosion combinatoire du nombre de variables et de contraintes.


 bien que le théorème des quatre couleurs fournisse une garantie théorique, l'application pratique de la coloration de cartes complexes, telles que celle des États-Unis, nécessite l'utilisation d'algorithmes efficaces comme DSATUR. De plus, la formulation de la problématique en termes de CSP offre une perspective structurée pour aborder la coloration de graphes, bien que des défis computationnels subsistent.
